```python
import numpy as np  
import matplotlib.pyplot as plt  

x=np.linspace(0,10,10)

y1=2*x+1
y2=np.cos(x)

plt.plot(x,y1,label="y=2x+1",marker="o",color="blue")
plt.plot(x,y2,label="cos(x)",marker="x",color="orange")

plt.title("Straight line and Cosine")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

```python
import pandas as pd

data = pd.Series(["100", "200", "python", "300.12", "400"])

print("Original Data Series:")
print(data)
print("dtype:", data.dtype)

data_numeric = pd.to_numeric(data, errors='coerce')

print("\nData Series after type conversion:")
print(data_numeric)
print("dtype:", data_numeric.dtype)
```
