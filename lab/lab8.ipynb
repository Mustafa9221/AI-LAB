```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

house_size = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)  
house_price = np.array([200, 250, 300, 350, 400]) 
```

```python
plt.scatter(house_size, house_price, color='blue')
plt.title("House Size vs Price")
plt.xlabel("Size (sq ft)")
plt.ylabel("Price (in $1000s)")
plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
model.fit(house_size, house_price)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope (m): {slope}")
print(f"Intercept (b): {intercept}")
```

```python
predicted_price = model.predict([[1800]])
print(f"Predicted price for 1800 sq ft: ${predicted_price[0]:.2f}")
```

```python
line_x = np.array([1000, 3000]).reshape(-1, 1)
line_y = model.predict(line_x)
plt.scatter(house_size, house_price, color='blue', label='Actual data')
plt.plot(line_x, line_y, color='red', label='Regression line')
plt.title("House Size vs Price with Regression Line")
plt.xlabel("Size (sq ft)")
plt.ylabel("Price (in $1000s)")
plt.legend()
plt.grid(True)
plt.show()
```
